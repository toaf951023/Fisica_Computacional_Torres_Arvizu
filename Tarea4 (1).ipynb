```julia
function newton(f::Function, df::Function, w0::Float64) #se crea una funcion que tome los parametros que reuiera la ecuacion par solucionarse  
    w1=w0
    w2=0.0
    tolerancia=0.000000000001
    error=0.001
  
    raices=[]
    while error >  tolerancia
        w2=w1-(f(w1)/df(w1))
        error=abs(w2-w1)
        w1=w2
        
        
        push!(raices,w1)
    end 
    return raices
end
```

Creare una funciòn que me guarde el error relativo con cada iteración 

```julia
function newtonerror(f::Function, df::Function, w0::Float64,s::Float64)
         #se crea una funcion que tome los parametros que reuiera la ecuacion par solucionarse  
    w1=w0
    w2=0.0
    tolerancia=0.000000000001
    error=0.001
    error2=0.0
    
    
    errores=[]
    while error >  tolerancia
        w2=w1-(f(w1)/df(w1))
        error=abs(w2-w1)
        error2=abs(w2-s)/s
        w1=w2
        
        
        push!(errores,error2)
    end 
    return errores
end
```

Creamos dos funciones que recorran el arreglo de soluciones y agregue a un nuevo arreglo las solucions que no repetidas.

```julia
function b(x)
n=length(x)
    t=[]
    push!(t,x[1])
    for i in 1:n
       if x[i] ∉ t
          push!(t,x[i])  
    
       end 
    end
   return t 
    
 end 
```

```julia
function buscadoraderaices(x)
n=length(x)
    raicesrepetidas=[]
    y=0.0
    epsilon=0.00000001
    push!(raicesrepetidas,x[1])
    for i in 2:n
         y=x[i]
        v=x[i-1]
          if abs(y-v)>epsilon
         
            push!(raicesrepetidas,y)
        end   
      end 
    return raicesrepetidas
end 
```
