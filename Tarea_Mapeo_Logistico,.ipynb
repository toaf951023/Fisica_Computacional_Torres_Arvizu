```julia
using PyPlot
```

```julia
function mapeologistico(r, x0)
    
    xs=[]
    for i in 1:length(r)
        x01=x0
        rl=r[i]
        for j in  1:10^4
            x01=rl*x01*(1-x01)
                  
    end
      push!(xs, x01)
    end
    return  xs
end
```

```julia
function iterar(r,x0i)
      ri=r
    for l in 1:length(x0i)
        plot(ri, mapeologistico(ri, x0i[l]))
        title("Diagrama de bifurcaci√≥n")
        xlabel(L"r")
        ylabel(L"x0")
    end
end  
```

```julia
iterar(0:0.01:7,[rand() for i in 1:12] )
```
