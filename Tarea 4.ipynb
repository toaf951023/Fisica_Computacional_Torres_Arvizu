Graficar como convergen  4  condiciones iniciales a las raiz real de  x3−x2−x−1 utilizando el método de Newton. Ésto básicamente es gráficar como se mueven los puntos  xk

La función de metodo de Newton que programe en la tarea 2 toma como argumento la función, su derivadas y punto fijo, asi que la utilizaré. 

```julia
function newton(f::Function, df::Function, w0::Complex{Float64}) #se crea una funcion que tome los parametros que reuiera la ecuacion par solucionarse  
    w1=w0
    w2=0.0
    tolerancia=0.000000000001
    error=0.001
   m=0
    raices=[]
    while error >  tolerancia
        w2=w1-(f(w1)/df(w1))
        error=abs(w2-w1)
        w1=w2
        m+=1
        println(m," ","iteración")
        push!(raices,w1)
    end 
    return raices
end
```

```julia
function newtonerror(f::Function, df::Function, w0::Float64,s::Float64)
         #se crea una funcion que tome los parametros que reuiera la ecuacion par solucionarse  
    w1=w0
    w2=0.0
    tolerancia=0.000000000001
    error=0.001
    error2=0.0
     m=0
    errores=[]
    while error >  tolerancia
        w2=w1-(f(w1)/df(w1))
        error=abs(w2-w1)
        error2=abs(w2-s)/s
        w1=w2
        m+=1
        println(m," ","iteración")
        push!(errores,error2)
    end 
    return errores
end
```

```julia
function newtonintervalo(f::Function, df::Function, w0::Array) #se crea una funcion que tome los parametros que reuiera la ecuacion par solucionarse
    raices=[]
    n=lenght(w0)
    w1=0
    w2=0.0
    tolerancia=0.000000000001
    error=0.001
   m=0
    for i in n
        w1=w0[i]
       while error >  tolerancia
          w2=w1-(f(w1)/df(w1))
           error=abs(w2-w1)
          w1=w2
          m+=1
          println(m," ","iteración")
         push!(raices,w1)
      end 
    end 
    return raices
end
```
