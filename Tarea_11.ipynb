## Solución 


```julia
using PyPlot
```

```julia


function eulerFunction(f::Function,time0,timef,y0,delta)
    #Se declaran dos arreglos a los cuales iremos agregando los valores calculados de t y y
    y=[]
    t=[]
    
    push!(t,time0)
    push!(y,y0)
  
    while(t[length(t)]<timef)
        
        
        push!(t,t[length(t)]+delta)
        push!(y,y[length(y)]+f(t[length(t)-1],y[length(y)])*delta)
        
    end
    title("Solución")
  
    scatter(t, y, color = "blue")
    grid()
    
   ylabel("y")
    
    xlabel("x") 

end
```

```julia
function eulercompFunction(f::Function,f2::Function,time0,timef,y0,delta)
    #Se declaran dos arreglos a los cuales iremos agregando los valores calculados de t y y
    y=[]
    t=[]
    solucion=[]
    push!(t,time0)
    push!(y,y0)
    push!(solucion, f2(time0))
    while(t[length(t)]<timef)
        
       
        push!(t,t[length(t)]+delta)
         push!(solucion, f2(t[length(t)]))
        push!(y,y[length(y)]+f(t[length(t)-1],y[length(y)])*delta)
        
    end
    title("Solución con el Método de Euler vs solución exacta vs ")
     plot(t,solucion, color="red", label="Solución exacta")
    legend()
    scatter(t, y, color = "blue",label="Solución método de Euler")
    legend()
    grid()
    
   ylabel("y")
    
    xlabel("x") 

end
```

## Ejercicio 2
¿Qué sistema físico está modelado por $\ddot{x} =
-kx$?
Escriba la solución exacta en terminos de las condiciones iniciales $x(0)$ y $x'(0)$.

## Solución 
El fenomeno físico que representa esta ecuación es el oscilador armónico simple, el cual puede ser un resorte que se estira sin fricciòn.
Podemos resolver las ecuaación diferencial proponiendo una solución de la forma $x(t)=e^{\lambda t}$, si se sustiye en la ecuación diferencial se obtiene,  
$e^{\lambda t}(\lambda^2+k)=0$, como $k>0$,el polinomio tiene las soluciones $\lambda=\pm\sqrt{k}i$.
Si sustituimos en la solución propuesta,  
  $x(t)=c_1e^{ \sqrt{k}it}+c_2e^{-\sqrt{k}it}$, por la identidad de Euler $e^{\theta i}=\cos(\theta )+i\sin(\theta )$ y acomodando las constantes, se obtiene 
    $x(t)=k_1\cos(\sqrt{k}t )+k_2\sin(\sqrt{k}t )$, aplicando las condiciones inciales  $x(0)$ y $x'(0)$, donde $x(0)=x_0$ y $x'(0)=0$,  
     $x(t)=x_0\cos(\sqrt{k}t)$, a esto se le puede agregar una fase $\phi$,y  finalmente la solución es $x(t)=x_0\cos(\sqrt{k}t+\phi)$.

## Ejercicio 3
Haz $u = \dot{x}(t)$ y calcula la derivada de $u$ ¿qué obtienes y como se relaciona con el sistema original?
Si llevaste a cabo el ejercicio anterior, te darás cuenta que terminas con un sistema acoplado de ecuaciones diferenciales:
$$
\dot{u} = -kx, \dot{x} = u,
$$
Con este sistema acoplado puedes utilizar el método de Euler en su forma vectorial: 
$$\frac{d\vec x}{dt}=\vec f(\vec x,t),$$ donde $\vec x=(x,u)$, y para este caso es facil ver que $\vec f(\vec x,t)=(u,-kx).$

## Solución

Si $u = \dot{x}(t)$, entonces su dervada $\dot{u}=\ddot{x}(t)$, susutiyuyendo en el sistema original, 
$\dot{u}= -kx$, entonces  se puede aplicar el método de Euler para encontrar primeramente $u$ y a partir de ésta $x(t)$. Si generalizamos el método de Euler para orden superior, obtenemos la siguiente expresión. 

$$\frac{d\vec x}{dt}=\vec f(\vec x,t),$$ donde $\vec x=(x,u_1, u_2,..., u_n)$  y  $ u_n=\frac{d^{n-1} x}{dt^{n-1}}$.
entonces el método de Euler vectorial es 
$$\vec y_{n+1} =\vec y_n+hf(x_n,\vec y_n) $$ para el caso del oscilador armónico simple  $\vec f(\vec x,t)=(u,-kx).$

Obten la solución del sistema $\ddot{x} = -kx(t)$ utilizando lo escrito arriba y el método de Euler.

## Solución
Se programarà el método de Euler de segundo orden, esto implica que se utilizará un sistema de dos ecuaciones diferenciales, donde se obtendrá primeramente, $x'(t)$,y por consiguiente $x(t)$ . Esta función será muy similar a la del ejercicio 1 pero,  las entradas de la función se mofican, ahora  las cuales recibirán dos funciones ( $x'(t)$ y $x(t)$), las condiciones inciales, el intervalo (punto incial y final) y el tamaño de paso con el cual se harà el metodo de Euler, la funcion nos devolverá las gráficas de $x'(t)$ y $x(t)$.




```julia
function euler2orden(f::Function,f2::Function,time0,timef,y10,y20,delta)
    y2=[]
    y1=[]
    t=[]
    push!(t,time0)
    push!(y1,y10)
     push!(y2,y20)
    while(t[length(t)]<timef)
           push!(t,t[length(t)]+delta)
        push!(y1,y1[length(y1)]+f(t[length(t)-1],y1[length(y1)],y2[length(y2)])*delta)
         push!(y2,y2[length(y2)]+f2(t[length(t)-1],y1[length(y1)-1],y2[length(y2)])*delta)
    end
     figure()
    subplot(2,1,1)
    plot(t, y2, color = "blue")
    legend()
    title("Gráfica de la función")
    ylabel("y1")
    xlabel("t")
    grid()
    
    
    subplot(2,1,2)
    plot(t,y1, color="red")
    legend()
    ylabel("y2")
    title("Gráfica de la derivada de la función")
    grid()
    println("the end")
    
end
```

Se prueba la función con $k=1$ y  $x'(t)=0$ y $x(t)=1$, con $t\in(0,20)$.

```julia
euler2orden((t,y1,y2)->y2,(t,y1,y2)->-1.0*y1, 0.0,20.0,0.0,1.0,0.01)
```
